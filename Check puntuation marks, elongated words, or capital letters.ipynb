#Check puntuation marks, elongated words, or capital letters in sentence
The code below is used to check whether in a sentence there are punctuation marks, elongated words, or capital letters.

```python
import re
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
```

```python
def punctuation_mark(text):
  """
  If in a sentence there is at least one question mark or exclamation mark, 
  return 1. Otherwise return 0.
  """
  for word in nltk.word_tokenize(str(text)):
    if word == '?' or word == '!':
      return 1
  return 0

def capital_word(text):
  """
  If a word is found in a sentence that is written in uppercase and 
  the word is longer than 1 letter, return 1. 
  Otherwise return 0.
  """
  for word in nltk.word_tokenize(str(text)):
    if word.isupper() and len(word) > 1:
      return 1
  return 0

def elongated_word(text):
  """
  If one elongated word is found in a sentence, return 1. 
  Otherwise return 0.
  """
  elong = re.compile("([a-zA-Z])\\1{2,}")
  for word in nltk.word_tokenize(str(text)):
    if elong.search(word):
      return 1
  return 0
```

```python
text = "Woooow SANGAT menarik!"
print("Punctuation marks =", str(punctuation_mark(text)))
print("Capital word =", str(capital_word(text)))
print("Elongated word =", str(elongated_word(text)))
```

"Woooow SANGAT menarik!" contains:
*   Woooow is elongated word
*   SANGAT is capital word
*   ! is punctuation mark



```python
text2 = "Angka ini sendiri, menurut Georgieva, jauh lebih rendah dibandingkan negara-negara lain di dunia, sehingga ia merasa bangga dengan pencapaian Indonesia tersebut."
print("Punctuation marks =", str(punctuation_mark(text2)))
print("Capital word =", str(capital_word(text2)))
print("Elongated word =", str(elongated_word(text2)))
```
